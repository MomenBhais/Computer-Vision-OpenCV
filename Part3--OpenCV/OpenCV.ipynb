```python
import cv2 
```

```python
#read image
img = cv2.imread(r"D:\Work\Route\HTI\S15\OpenCV\cats.jpg")
```

```python
#show image
cv2.imshow('Cats', img) #RGB
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img) #BGR
plt.axis('off')
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR ---> RGB
plt.imshow(img)
plt.show()
```

```python
#convert image to gray image
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray Image', gray_image)
cv2.waitKey(0)
```

```python
#convert image to HSV Image ---> Hue, Saturation, Value
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV Image', hsv_image)
cv2.waitKey(0)
```

```python
#shape of image
img.shape
```

```python
#resize image
resized_img = cv2.resize(img, (500, 500), interpolation = cv2.INTER_CUBIC)
cv2.imshow('resized Image', resized_img)
cv2.waitKey(0)
```

```python
#CROP ON selected area
crop_image = img[20:220, 150:293] #height, width
cv2.imshow('crop Image', crop_image)
cv2.waitKey(0)
```

```python
#read video
capture = cv2.VideoCapture(r"D:\Work\Route\HTI\S15\OpenCV\dog.mp4")
```

```python
capture.read()
```

```python
while True:
    isTrue, frame = capture.read()
    if isTrue:
        cv2.imshow('Video', frame)
        if cv2.waitKey(0) & 0xFF == ord('d'):
            break
        else:
            break
capture.release()
cv2.destroyAllWindows()
```

```python
blur_image = cv2.GaussianBlur(img, (9,9), 0)
cv2.imshow('blue', blur_image)
cv2.waitKey(0)
```

```python
cany_image = cv2.Canny(blur_image, 75,150)
cv2.imshow('cany_image', cany_image)
cv2.waitKey(0)
```

```python
import numpy as np
blank = np.zeros((500,500,3),dtype = 'uint8')
cv2.imshow('blank IMG', blank)
cv2.waitKey(0)
```

```python
blank[200:300,300:400] = 0,0,255
cv2.imshow('Color IMG', blank)
cv2.waitKey(0)
```

```python
cv2.rectangle(blank,(blank.shape[1]//2,blank.shape[0]//2),(blank.shape[1],blank.shape[0]),(0,255,0),thickness= 10)
cv2.imshow('regtangle IMG', blank)
cv2.waitKey(0)
```

```python
cv2.circle(blank,(blank.shape[1]//2,blank.shape[0]//2),100,(200,0,0),thickness= 5)
cv2.imshow('regtangle IMG', blank)
cv2.waitKey(0)
```

```python
cv2.putText(blank,'AI & Machine Learning',(0,200),cv2.FONT_HERSHEY_TRIPLEX,1,(255,255,255),3)
cv2.imshow('text IMG', blank)
cv2.waitKey(0)
```

```python
contours , hierarchies = cv2.findContours(cany_image,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
blank = np.zeros((500,500,3),dtype = 'uint8')
cv2.drawContours(blank,contours,-1,(255,255,255),-1)
cv2.imshow('text IMG', blank)
cv2.waitKey(0)
```
